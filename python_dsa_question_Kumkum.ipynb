Question 1. Reverse an Array

```python
def reverse_array(arr:list)->list:
    n=len(arr)
    reversed_list=[]
    for i in range(n-1,-1,-1):
        reversed_list.append(arr[i])
    return reversed_list

print(f"Input: [1,2,3],Output:{reverse_array([1,2,3])}")
```
